# Router subspace ablation – plot results

Load a `results_ablation.json` (or any output from the ablation run) and plot loss and Δ loss by variant (svd, orthogonal, random).

```python
import sys
from pathlib import Path

# Project root: directory that contains "src"
ROOT = Path.cwd() if (Path.cwd() / "src").exists() else Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.experiments.router_subspace_ablation.plotter import load_results, plot_ablation_results
```

```python
# Path to your results JSON (default output from run is results_ablation.json)
RESULTS_PATH = ROOT / "results_ablation.json"  # or absolute path, e.g. "/path/to/results_ablation.json"

results = load_results(RESULTS_PATH)
print("Baseline loss:", results.get("baseline_loss"))
print("Variants:", list(results.get("results", {}).keys()))
```

```python
%matplotlib inline

fig = plot_ablation_results(results, show_baseline=True)
fig.show()
```

```python
# Optional: save figure
OUTPUT_FIG = ROOT / "ablation_plots.png"
plot_ablation_results(results, save_path=OUTPUT_FIG)
print(f"Saved to {OUTPUT_FIG}")
```
